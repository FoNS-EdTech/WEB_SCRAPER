```python
#Import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Define URL
url = 'https://imperial.cloud.panopto.eu/Panopto/Pages/Analytics.aspx#folderId=637c507d-8415-4328-b882-36d46997f3a7&dateFilterItem=4&breakoutTableType=2'
# Ask hosting server to fetch url
requests.get(url)
```

```python
pages = requests.get(url)
pages.text
```

```python
# parser-lxml = Change html to Python friendly format
soup = BeautifulSoup(pages.text, 'lxml')
soup
```

```python
# Obtain information from tag <table>
tabli = soup.find('table' div="sortable-table")
print(tabli)
```
